```python
# Cargamos dataset 
df = spark.read.csv("/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv", header=True, inferSchema=True)
display(df.limit(5))
```

```python
# Filtramos el dataset, filtramos diamantes de alta calidadd y renombro la columna
df_filtrado = df.filter(df.cut == "Ideal").withColumnRenamed("price", "precio")
display(df_filtrado)
```

```python
# Realizamos consultas con Spark SQL
# Creamos una vista temporal para utilizar SQL
df_filtrado.createOrReplaceTempView("tabla_diamantes")

# Mostramos el resultado de la consulta
resultado_sql = spark.sql("SELECT color, AVG(precio) as promedio FROM tabla_diamantes GROUP BY color")
display(resultado_sql)
```
